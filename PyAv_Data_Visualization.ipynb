# Matplotlib


## Gráfico Básico

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Dados/imigrantes_canada.csv')
```

```python
df
```

```python
df.info()
```

```python
df.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]
```

```python
brasil
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
brasil_dict
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2015'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Imigrantes')
plt.show() ## abaixo disso nada é executado, tem sempre que ser o último comando
```

## Prática


```python
df_comparacao = df.loc[['Brasil', 'Argentina'], anos]
df_comparacao
```

```python
df_comparacao = df_comparacao.T
```

```python
df_comparacao
```

```python
plt.plot(df_comparacao['Brasil'],label = 'Brasil')
plt.plot(df_comparacao['Argentina'],label ='Argentina')
plt.title('Imigração do Brasil e Argentina para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2015'])
plt.legend()
plt.show()
```

## Figuras (Subplot)

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
import numpy as np
fig, axs = plt.subplots(1,2, figsize=(15,5))
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()


brasil_q1 = np.percentile(df_comparacao['Brasil'], 25)
brasil_q3 = np.percentile(df_comparacao['Brasil'], 75)
brasil_mediana =  np.percentile(df_comparacao['Brasil'], 50)
axs[1].boxplot(df_comparacao)
axs[1].set_title('Comparação entre Brasil e Argentina')
axs[1].set_xlabel('Brasil / Argentina')
axs[1].set_ylabel('Imigrantes')


axs[1].annotate(f'Q1: {brasil_q1:.2f}', xy=(1, brasil_q1), xytext=(1.2, brasil_q1))
axs[1].annotate(f'Mediana: {brasil_mediana:.2f}', xy=(1, brasil_mediana), xytext=(1.2, brasil_mediana))
axs[1].annotate(f'Q3: {brasil_q3:.2f}', xy=(1, brasil_q3), xytext=(1.2, brasil_q3))

axs[1].grid()
plt.show()
```
