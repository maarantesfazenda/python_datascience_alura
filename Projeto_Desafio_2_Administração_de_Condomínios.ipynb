```python
import pandas as pd
df = pd.read_json('/content/dados_locacao_imoveis.json')
df
```

```python
df = pd.json_normalize(df['dados_locacao'])
df
```

```python
colunas = list(df.columns)
colunas
```

```python
df = df.explode(colunas[1:])
df
```

```python
df.reset_index(inplace=True, drop=True)
df
```

```python
df['valor_aluguel'] = df['valor_aluguel'].apply(lambda x: x.replace('$', '').replace('.', '').replace(',', '.').replace('reais','').strip())
```

```python
df
```

```python
import numpy as np
df['valor_aluguel'] = df['valor_aluguel'].astype(np.float64)
```
