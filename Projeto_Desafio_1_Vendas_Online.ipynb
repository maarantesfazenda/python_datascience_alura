```python
import pandas as pd
```

```python
df = pd.read_json('/content/drive/MyDrive/Dados/dados_vendas_clientes.json')
df
```

```python
df = pd.json_normalize(df['dados_vendas'])
df
```

```python
colunas = list(df.columns)
```

```python
colunas
```

```python
df = df.explode(colunas[1:])
```

```python
df.reset_index(inplace=True, drop=True)
```

```python
df.info()
```

```python
df
```

```python
df['Valor da compra'] = df['Valor da compra'].apply(lambda x: x.replace('R$','').replace(',','.').strip())
```

```python
import numpy as np
df['Valor da compra'] = df['Valor da compra'].astype(np.float64)
```

```python
df
```

```python
df['Cliente'] = df['Cliente'].str.lower()
df
```

```python
df['Cliente'] = df['Cliente'].str.replace('[^a-zA-Z\'\-]',' ', regex=True)
df
```

```python
df['Cliente'] = df['Cliente'].str.split()
```

```python
df
```

```python
df.info()
```

```python
df['Data de Venda'] = pd.to_datetime(df['Data de venda'])
```
