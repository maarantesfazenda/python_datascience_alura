```python
import pandas as pd
import json
df = pd.read_json('/content/drive/MyDrive/Dados/cursos_cadastrados.json')
df.head()
```

```python
#usando a biblioteca json para normalizar
with open('/content/drive/MyDrive/Dados/cursos_cadastrados.json') as file:
  json.bruto = json.load(file)
```

```python
df = pd.json_normalize(json.bruto)
```

```python
df.isnull().sum().sum()
```

```python
df[df.isna().any(axis=1)]
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df[df['instrutor.nome'] == ""]
```

```python
df[df['data_conclusao'] == ""]
```

```python
# Substitui strings vazias por valores nulos
df.replace('', pd.NA, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
# Converte a concluintes para o tipo inteiro
df['concluintes'] = df['concluintes'].astype(int)

# Converte a coluna data_inicio e data_conclusao para o tipo datetime
df['data_inicio'] = pd.to_datetime(df['data_inicio'])
df['data_conclusao'] = pd.to_datetime(df['data_conclusao'])

# Convertendo a coluna preço para o tipo float
df['preco'] = df['preco'].astype(float)
```
